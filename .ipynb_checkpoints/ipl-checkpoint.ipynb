```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/Users/harishkumar/Downloads/IPL Player Stats - 2016 till 2019.csv')
df.head()
```

```python
df_2019 = df[df['Tournament'] == 'IPL 2019']
df_2019.head()
```

```python
df.replace({'-': int(0)}, inplace=True)
df.head()
```

```python
df_2019['Fantasy']=(df_2019['Runds Scored'])*2+(df_2019['Wickets Taken'])*20
```

The Fantasy Points are calculated by using the given formula:

Points = Runs Scored * 2 + Wickets Taken * 20

```python
df.info()
```

```python
df[['Runds Scored','Wickets Taken' ]] = df[['Runds Scored','Wickets Taken']].apply(pd.to_numeric)
```

```python
df['Fantasy']=(df['Runds Scored'])*2+(df['Wickets Taken'])*20
```

```python
df.head()
```

```python
a=df[(df["Player"] == 'RG Sharma')]
a
```

```python
ax = a.plot.line(x='Tournament', y='Fantasy', rot=0)
```

```python
grouped = df.groupby('Tournament')
year_avg= (grouped['Fantasy'].agg(np.mean))
print(year_avg)
```

```python
ax = year_avg.plot.line(x='Tournament', y='mean', rot=0,label='Teams')
```

```python
teams = df.groupby('Team')

csk=teams.get_group('Mumbai Indians')
grouped = csk.groupby('Tournament')
csk_year_avg= (grouped['Fantasy'].agg(np.mean))
print(csk_year_avg)
```

```python
ax = csk_year_avg.plot.line(x='Tournament', y='Fantasy', rot=0,label='MI')
```

```python
ax = a.plot.line(x='Tournament', y='Fantasy', rot=0,label='Rohit')
ax = year_avg.plot.line(x='Tournament', y='mean', rot=0,label='Teams')
ax = csk_year_avg.plot.line(x='Tournament', y='Fantasy', rot=0,label='MI')
ax.legend(["Rohit","Teams","MI"])
```

```python
df[['Overs Bowled','Bowling Economy Rate' ]] = df[['Overs Bowled','Bowling Economy Rate']].apply(pd.to_numeric)
df['Bowling Average'] = df['Bowling Average'].astype(int)
```

```python
a=df[(df["Team"] =="Mumbai Indians")]
a=a[(a["Overs Bowled"] >=20)]
a
```

```python
a=a.nsmallest(6, 'Bowling Average')
a
```

```python
ax = a.plot.barh(x='Player', y='Bowling Economy Rate', rot=0)
ax = a.plot.line(x='Player', y='Wickets Taken', rot=0,label='Wickets Taken')
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
color = 'tab:blue'
ax1.set_title('SUPER OVER', fontsize=16)
ax1.set_xlabel('Player', fontsize=16)
ax1.set_ylabel('Avg Temp', fontsize=16)
ax1 = sns.barplot(x='Player', y='Bowling Economy Rate', data = a, palette='summer')
ax1.tick_params(axis='y')
#specify we want to share the same x-axis
ax2 = ax1.twinx()
color = 'tab:red'
#line plot creation
ax2.set_ylabel('Avg Percipitation %', fontsize=16)
ax2 = sns.lineplot(x='Player', y='Wickets Taken', data = a, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
#show plot
plt.show()
```

```python
df['Batting Innings'] = df['Batting Innings'].astype(int)
a=df[(df["Team"] =="Mumbai Indians")]
a=a[(a["Batting Innings"] >=5) & (a["4s"]+a["6s"] >=25)]
a
```

```python
df['4s'] = df['4s'].astype(int)
df['6s'] = df['6s'].astype(int)
df['Balls Faced'] = df['Balls Faced'].astype(int)
a['Balls per boundary']=(a['4s']+a['6s'])/(df['Balls Faced'])*100
a
```

```python
a['Balls per boundary'] = a['Balls per boundary'].astype(int)
```
